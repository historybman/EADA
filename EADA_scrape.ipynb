# A test of webscraping
This is an experiment to get some text into a database from the Early Americas Digital Archive

```python
import urllib.request
html = urllib.request.urlopen('http://eada.lib.umd.edu/?s=&post_type=eada_entry&tax_input%5Bgeographic%5D%5B%5D=96&action=search', data = None)

```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
import re
new_list = []
for link in soup.find_all('a'):
    test = link.get('href') 
    if test not in new_list:
        new_list.append(test)

print(new_list)
    
```

```python
new_list2 = []
for i in new_list:
    if re.match('^http:', i):
        new_list2.append(i)
    else:
        continue

print(new_list2)
```

## Next steps:
What I have now is a list of urls that I can use urllib to open and then parse with beautiful soup using the get_text() function.

```python
def save_text(url):
    soup = BeautifulSoup(url, 'html.parser')
    text = soup.get_text()
    print(text[:200])
```

```python
import time

for cycle in range(len(new_list2)):
    html = urllib.request.urlopen(new_list2[cycle], data=None)
    web_text = save_text(html)
    print(web_text)
    time.sleep(1)
    print('cycle {} is done.'.format(cycle))
```
